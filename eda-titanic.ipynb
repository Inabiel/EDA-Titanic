```python
%pip install matplotlib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Import pandas dataframe from url
URL = "https://raw.githubusercontent.com/datasciencedojo/datasets/refs/heads/master/titanic.csv"
df = pd.read_csv(URL)

df
```

```python
#exploratory data analysis
#get the shape of the dataframe, it is 891 rows and 12 columns
print(df.shape)

#get first n data of dataframe
print(df.head(10))
```

```python
#get the column names of the dataframe
#column is ['PassengerId', 'Survived', 'Pclass', 'Name', 'Sex', 'Age', 'SibSp', 'Parch', 'Ticket', 'Fare', 'Cabin', 'Embarked']
print(df.columns.tolist())
```

```python

#get row that has null values
nullValues = df.columns[df.isnull().any()]
print(', '.join(nullValues))   
```

```python
#get total percentage of survived passenger
survivedPassengerCount = df['Survived'] == 1
print(f"{survivedPassengerCount.sum()/len(df)*100:.2f}% of passengers survived")
```

```python
#get mean of age of passengers who survived where class is 1 and is female
survivedPassengerAge = df[(df['Survived'] == 1) & (df['Pclass'] == 1) & (df['Sex'] == 'female')]['Age'] 
print(f"The mean age of passengers who survived where class is wk1 is {survivedPassengerAge.mean().__round__(1)} years old") 
```

```python
#get class of passenger with the highest survival rate
#visualize survival rate
hasil = df.groupby('Pclass').agg(
    total=('PassengerId','count'),
    selamat=('Survived','mean'),
).round(3)
print(hasil)

survivalRateByClass = df.groupby('Pclass')['Survived'].mean()
highestSurvivalRateClass = survivalRateByClass.idxmax()
print(f"The class of passenger with the highest survival rate is class {highestSurvivalRateClass}")
```

```python
#result of groupBy pclass survival mean
meanOfGroupByPClassSurvival = df.groupby('Pclass')['Survived'].mean()
print(meanOfGroupByPClassSurvival)
```

```python
#independent exploratory data analysis
getMaximumFareOfAllClass = df.groupby('Pclass')['Fare'].max()
getMinimumFareOfAllClass = df.groupby('Pclass')['Fare'].min()

getWhereFareisZero = df.loc[df['Fare'] == 0.0]
print(getMaximumFareOfAllClass)
print(getMinimumFareOfAllClass)
print(getWhereFareisZero)
```

```python
#get spread of ticket fare by class
df.plot.scatter(x='Age', y='Fare')
plt.show()
```
